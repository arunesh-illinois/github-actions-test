## üìö 2. Import Libraries

```python
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

## üìç 3. Create Sample Point Data (GeoDataFrame)

```python
data = {
    'name': ['Point A', 'Point B', 'Point C'],
    'latitude': [40.1106, 40.1138, 40.1042],
    'longitude': [-88.2073, -88.2272, -88.1986]
}
geometry = [Point(xy) for xy in zip(data['longitude'], data['latitude'])]
gdf_points = gpd.GeoDataFrame(data, geometry=geometry, crs='EPSG:4326')
gdf_points
```

## üåç 4. Load World Map Data

```python
import geopandas as gpd
import geodatasets

# Load the world countries dataset
world = gpd.read_file(geodatasets.get_path('naturalearth.land'))
world.head()
```

## üñºÔ∏è 5. Plot Points on Static World Map (Matplotlib)

```python
fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax, color='lightgrey')
gdf_points.plot(ax=ax, color='red', markersize=100)
plt.title('Sample Points on World Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
